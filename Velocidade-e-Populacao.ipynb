```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Tmax = 50

dt = 0.01
nt = round(Tmax/dt)
time = np.linspace(0, Tmax, nt)
```

```python
def velocity(a,b,Vo):
    velocity = np.zeros(nt)
    velocity[0] = Vo
    
    for i in range(1, nt): velocity[i] = velocity[i-1]*(1 - b*dt) + a*dt    
    
    return velocity


def number(a,b,No): 
    Number = np.zeros(nt)
    Number[0] = No
    
    for i in range(1, nt): Number[i] = Number[i-1]*(1 + a*dt - b*Number[i-1]*dt)
    
    return Number


```

## Velocidade

```python
Blist = [0.1,1,10]

for b in Blist:
    v = velocity(10, b, 0)
    plt.plot(time, v)
```

```python
Tmax = 50
v = velocity(10, 0.5, 0)
plt.plot(time, v)
```

## Crescimento Populacional

```python
Tmax = 1.5
No = 5

N = number(10,4,No)
plt.plot(time, N, 'green')

N = number(10,3,No)
plt.plot(time, N, 'red')

N = number(10,2,No)
plt.plot(time, N, 'yellow')
```

```python
Tmax = 1.5
No = 1000

N = number(10,0.01,No)
plt.plot(time, N, 'green')

N = number(10,0.005,No)
plt.plot(time, N, 'red')

N = number(10,0.015,No)
plt.plot(time, N, 'yellow')
```
