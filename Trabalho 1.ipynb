```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y1(x):
    return x**3 + x**2 - 10*x + 10

def y2(x):
    return np.sin(x)/x

def dy1(x):
    return 3*x**2 + 2*x -10

def dy2(x):
    return (x*np.cos(x) - np.sin(x))/x

def dy3(x):
    return "nada ainda"
```

```python
def dy(f,Xmin, Xmax, dx):
    steps = round((Xmax - Xmin)/dx)

    yaxis = np.zeros(steps)
    xaxis = np.linspace(Xmin, Xmax, steps)

    for i in range(steps): yaxis[i] = f(xaxis[i])
        
    return xaxis, yaxis

```

```python
x, y = dy(y1,-5,5,0.1)
plt.grid(linewidth = 0.5)
plt.plot(x,y)
```
